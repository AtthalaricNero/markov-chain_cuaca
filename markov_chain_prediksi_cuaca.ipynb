```python
import requests
import pandas as pd
from datetime import datetime, timedelta

end_date = datetime.now().strftime('%Y-%m-%d')
start_date = (datetime.now() - timedelta(days=100)).strftime('%Y-%m-%d')

cuaca_url = f"https://archive-api.open-meteo.com/v1/archive?latitude=-7.642635&longitude=112.703293&start_date={start_date}&end_date={end_date}&daily=weathercode,temperature_2m_max,precipitation_sum&timezone=Asia%2FJakarta"

response = requests.get(cuaca_url)
data = response.json()

df = pd.DataFrame(data['daily'])

print(df)
```

```python
from math import inf
batas_rentang = [-0.1, 1, 10.0, float(inf)]
nama_state_cuaca = ['Cerah', 'Hujan', 'Hujan Deras']

df['State'] = pd.cut(df['precipitation_sum'], bins=batas_rentang, labels=nama_state_cuaca)

print("Dataframe dengan kolom state")
print(df)
```

```python
jumlah_state = df['State'].value_counts()

print(jumlah_state)
```

```python
df['Next State'] = df['State'].shift(-1)

transition_counts = pd.crosstab(df['State'], df['Next State'])
print(f"Jumlah kemunculan transisi: {transition_counts}")

transition_matrix = transition_counts.div(transition_counts.sum(axis=1), axis=0)
print(transition_matrix)
```
